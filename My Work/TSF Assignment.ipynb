### 1.2 Aggregation

Additive time series : $O_{t} = T_{t} + S_{t} + R_{t}$ 

$O_{t}$ is the output   
$T_{t}$ is the trend  
$S_{t}$ is the seasonality  
$R_{t}$ is the residual  
$_t$ is a variable representing a particular point in time

```python
additive = trend + seasonal + residual
```

```python
# plotting
plt.plot(time, additive, 'b.')
plt.title("Additive Time Series")
plt.xlabel("Minutes")
plt.ylabel("Demand");
```

we can clearly see the rising trend, little seasonal component and the magnitude of the series is not changing drastically -> Additive Time series

---

Multiplicative time series : $O_{t} = T_{t} * S_{t} * R_{t}$ 

```python
# let's just ignore the residual , inorder to view the pattern more apparently
multiplicative = trend * seasonal 
```

```python
# plotting
plt.plot(time, multiplicative, 'r-.')
plt.title("Multiplicative Time Series")
plt.xlabel("Minutes")
plt.ylabel("Demand");
```

There is a big fluctutaion in the magnitude of teh time series -> Multiplicative Time Series

## 2. Stationarity

What is Stationarity?
In order for time series data to be stationary, the data must exhibit four properties over time:
1. constant mean
2. constant variance
3. constant autocorrelation structure
4. no periodic component

### 2.1 Construct Stationary and Non- Stationary Plots

```python
# helper function for plotting
def run_sequence_plot(x, y, title, xlabel="time", ylabel="series"):
    plt.plot(x,y, 'b-')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
# stationary ts
time = np.arange(100)
stationary = np.random.normal(loc=2.0, scale =1.0, size=len(time))
run_sequence_plot(time, stationary, title="Stationary TS")
```

In the above plot, we can see there is a constant mean(2), there is no clear trend or seasonality and statisfies the four conditions. This is a example for Stationary time series

```python
# autocorrelation -> adding previous lag to the current data
start = 7.0

lagged = np.empty_like(time, dtype="float")
for t in time:
    lagged[t] = start + np.random.normal(loc=0, scale=1, size=1)
    start = lagged[t]
    
#plot
run_sequence_plot(time, lagged, title="NonStationary TS w/Lagged data")
```

The above plot is not stationary because, the current data is highly correlated with previous data.(Autocorrelation)

```python
#Other non stationary plots
# changing trend
trend = (time*3) + stationary  # adding stationary to show some noise
run_sequence_plot(time, trend, title="Nonstationary TS w/Trend")
```

The above plot is Not stationary because, mean is changing overtime

```python
#Heteroscedasticity
l1 = np.random.normal(loc=0,scale=1.0,size=50)
l2 = np.random.normal(loc=0,scale=10.0,size=50)
heteroscedasticity = np.append(l1,l2)
#plot
run_sequence_plot(time,heteroscedasticity,
                  title="NonStationary TS w/Heteroscedasticity")
```

The above plot is not stationary beacuse, the variance is changing overtime.

```python
#seasionality
seasionality = 10 + np.sin(time) * 10
run_sequence_plot(time, seasionality, title="Nonstationary TS w/Seasonality")
```

The above plot is not stationary because, there is a periodic component(Seasonality).

#### Augmented Dickey-Fuller Test
This is a statistical procedure to suss out whether a time series is stationary or not.Things to do

Frame Hypothesis
1. **Null hypothesis:** the series is nonstationary.
2. **Alternative hypothesis:** the series is stationary.

Like any statistical test you should set a significance level or threshold that determines whether you should accept or reject the null. 
> The value 0.05 is common but depends upons numerous factors.

```python
#library
from statsmodels.tsa.stattools import adfuller
#passing stationary TS
adf, pvalue, usedlag, nobs, critical_values, icbest = adfuller(stationary)
```

```python
print("1. ADF: {0};\n2. p-value: {1};".format(adf,pvalue))
print("3. Critical values:\n",critical_values)
```

Interperting the Results:
1. **adf** is the value of the test statistic. The more negative the value, the more confident we can be that the series is stationary. Here we see a value of -10.
2. **pvalue** is interpreted like any p-value. Once we set a threshold, we can compare this p-value to that threshold. Either we reject or fail to reject the null. Here **pvalue** is very close to zero (~$10^{-19}$) so we reject the null that this data is nonstationary.
3. the **critical_values** variable provides test statistic threholds for common significant levels. Here we see a test statistic of roughly -2.89 and lower is sufficient to reject the null using a significance level of 5%.

> From these reults, we can say that the TS is **stationary**.

### 3.2 Moving Average Smoothing 
The first step is to select a window size. We'll choose a size of three arbitrarily. So what happens is we compute the average for the first three values and store the result. We then slide the window over one spot and calculate the average of the next three values. We continue this process until we reach the final observed value.

```python
# Let's try smoothing with moving average
# fn taken from https://stackoverflow.com/questions/14313510/how-to-calculate-rolling-moving-average-using-numpy-scipy#14314054
def moving_average(observations, window=3, forecast=False):
    '''returns the smoothed version of an array of observations.'''
    cumulative_sum = np.cumsum(observations, dtype=float)
    cumulative_sum[window:] = cumulative_sum[window:] - cumulative_sum[:-window]
    if forecast:
        return np.insert(cumulative_sum[window - 1:] / window, 0, np.zeros(3))
    else:
        return cumulative_sum[window - 1:] / window
```

```python
#call the fn
smoothed_trend = moving_average(trend, window=3, forecast=False)
```

```python
run_sequence_plot(time, trend, title="Moving average Smoothing Nonstationary TS w/Trend")
plt.plot(time[1:], np.insert(smoothed_trend,0,0),"r.",label="MA Smoothing")
#for first datapoint we won't have the value, so we are slicing time, and creating an array with smoothed trend
plt.legend();
```

This Smoothing Technique could capture the trend propely, seems like an good estiamt for forcasting, but in Time series, we usually give more weightage to recent datapoints and give less weighatge to old datapoints. This method fails to do that, it gives equal weightage to all datapoints, inorder to overcome this we have **Exponential Smoothing**.

###  3.3 Exponential Smoothing
There are three key exponential smoothing techniques:

1. Single Exponential Smoothing - no trend or seasonality
2. Double Exponential Smoothing - captures trend
3. Triple Exponential Smoothing - captures trend & seasonality

#### 3.3.1 Single Exponential Smoothing
This method is useful if your data lacks trend and seasonality and you want to approximately track patterns in your data.

```python
# helper fn to calculate mse(mean squared error)
def mse(observations, estimates):
    diff = observations - estimates
    sq_diff = diff ** 2
    mse = sum(sq_diff)
    return mse
```

```python
#Let's compare Smiple Smoothing with Single Exp smoothing
#plot NS ts data
run_sequence_plot(time, NS, title="NonStationary TS")
```

```python
#Train/Test split
train = NS[:-5]
test = NS[-5:]

#Simple Smoothing
NS_avg = np.mean(NS)

NS_simple_smoothing = np.full(shape=len(test), fill_value=NS_avg, dtype=float)
#mse
simple_mse = mse(test, NS_simple_smoothing)

print("Predictions:",NS_simple_smoothing)
print("Test:", test)
print("MSE of Simple Smoothing:",simple_mse)
```

```python
plt.plot(time[:-5], train, 'b--', label="train")
plt.plot(time[-5:], test, color='orange', linestyle="--", label="test")
plt.plot(time[-5:], NS_simple_smoothing, 'r--', label="predictions")
plt.legend(loc='upper left')
plt.title("Simple Average Smoothing");
```

```python
#library
from statsmodels.tsa.api import SimpleExpSmoothing

ses = SimpleExpSmoothing(train).fit(optimized=True)
ses_preds = ses.forecast(len(test))
ses_mse = mse(test, ses_preds)
print("Predictions: ", ses_preds)
print("Test: ", test)
print("MSE of SES: ", ses_mse)
```

```python
plt.plot(time[:-5], train, 'b--', label="train")
plt.plot(time[-5:], test, color='orange', linestyle="--", label="test")
plt.plot(time[-5:], ses_preds, 'r--', label="predictions")
plt.legend(loc='upper left')
plt.title("Single Exponential Smoothing");
```

From the above plot it is clear that, Single Exponential Smoothing could not pick up trend and seasonality.

#### 3.3.2 Double Exponential Smoothing

```python
#library
from statsmodels.tsa.api import Holt

des = Holt(train).fit(optimized=True)
des_preds = des.forecast(len(test))
des_mse = mse(test, des_preds)
print("Predictions: ", des_preds)
print("Test: ", test)
print("MSE of DES: ", des_mse)
```

```python
plt.plot(time[:-5], train, 'b--', label="train")
plt.plot(time[-5:], test, color='orange', linestyle="--", label="test")
plt.plot(time[-5:], des_preds, 'r--', label="predictions")
plt.legend(loc='upper left')
plt.title("Double Exponential Smoothing")
```

From the above plot, we can see that Double Exponential Smoothing can pick up trend but fails to capture the seasonality

#### 3.3.3 Triple Exponential Smoothing

```python
#library
from statsmodels.tsa.api import ExponentialSmoothing

tes = ExponentialSmoothing(train,
                              trend="additive",
                              seasonal="additive",
                              seasonal_periods=13).fit(optimized=True)
tes_preds = tes.forecast(len(test))
tes_mse = mse(test, tes_preds)
print("Predictions: ", tes_preds)
print("Test: ",test)
print("MSE: ", tes_mse)
```

```python
plt.plot(time[:-5], train, 'b--', label="train")
plt.plot(time[-5:], test, color='orange', linestyle="--", label="test")
plt.plot(time[-5:], tes_preds, 'r--', label="predictions")
plt.legend(loc='upper left')
plt.title("Triple Exponential Smoothing");
```

Woww! we were able to capture trend, seasonality.

## 4. ARMA and ARIMA models
Your Time series data has to be **Stationary**

### 4.1 AutoRegressive (AR) model
**Autocorrelation:** a variable's correlation with itself at different lags.

AR models regress on actual past values 

This is the first order or **AR(1)** formula : $y_t = \beta_0 + \beta_1y_{t-1}+\epsilon_t$

The $\beta$'s are just like those in linear regression and $\epsilon$ is irreducible error.

A second order or **AR(2)** would look like this: $y_t = \beta_0 + \beta_1y_{t-1}+\beta_2y_{t-2}+\epsilon_t$

The pattern of adding another coefficient and another past term continues to whichever order you choose. 

```python
#Let's create our own AR(1) model
ar1 = np.zeros(100)

#first value mean=0, std.dev=2.5
ar1[0] = np.random.normal(loc=0,scale=2.5,size=1)

# Set every value thereafter as 0.7 * the last term plus a random error
for t in time[1:]:
    ar1[t] = (0.7 * ar1[t-1]) + np.random.normal(loc=0, scale=2.5, size=1)
```

```python
plt.title("AR(1)")
plt.fill_between(time, ar1);
```

```python
#library
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_pacf
#pacf plot, to verify the 0.7 correlation
fig = sm.tsa.graphics.plot_pacf(ar1, lags=range(1,15), alpha=0.05, title = 'pacf ar1')
```

```python
#let's use ARMA model and check the model parameter (phi value =0.7(approx) or not)
model = sm.tsa.ARMA(ar1, (1,0)).fit(trend="nc", disp=0)
model.params
```

```python
#Let's create our own AR(2) model
ar2 = np.zeros(100)

#first value mean=0, std.dev=2.5
ar2[0:2] = np.random.normal(loc=0,scale=2.5,size=1)

# Set every value thereafter as 0.3 * the last term plus a random error
for t in time[2:]:
    ar2[t] = (0.3 * ar2[t-1]) + (0.3 * ar2[t-2]) +np.random.normal(loc=0, scale=2.5, size=1)

plt.title("AR(2)")    
plt.fill_between(time, ar2)
```

```python
#pacf plot, to verify the 0.3 correlation
fig = sm.tsa.graphics.plot_pacf(ar2, lags=range(1,15), alpha=0.05, title = 'pacf ar2')
```

```python
##let's use ARMA model and check the model parameter (phi0 and phi1 value =0.3(approx) or not)
model = sm.tsa.ARMA(ar2, (2, 0)).fit(trend='nc', disp=0)
model.params
```

Perfect!!! we have achieved in creating a correlated data with lag1(0.7) and lag2(0.3) and verified it with ARMA model params.)

### 4.2 Moving Average(MA) Model
A MA model is defined by this equation: $y_t=c+e_t+_1e_{t1}+_2e_{t2}++_qe_{tq}$ where $e_t$ is white noise. The value $c$ is a constant value and the $\theta$'s are coefficients, not unlike those found in linear regression.

```python
#Let's create our own ma(1) model
ma1 = np.zeros(100)
err = np.random.normal(loc=0, scale=2.5, size=100)
ma1[0] += err[0]
#set future values to 0.6 times error of prior value plus the current error term
for t in time[1:]:
    ma1[t] = (0.6 * err[t-1]) + err[t]
    
plt.title("MA(1)")
plt.fill_between(time, ma1)
```

```python
fig = sm.tsa.graphics.plot_acf(ma1, lags=range(1,15), alpha=0.05,title = 'ma1 ACF')
```

```python
# find model params for generated sample, whether we get theta =0.6 (approx)
model = sm.tsa.ARMA(ma1, (0, 1)).fit(trend='nc', disp=0)
model.params
```

```python
#function taken from IBM machine learning course
def cross_validate(series,horizon,start,step_size,order = (1,0,0),seasonal_order = (0,0,0,0),trend=None):
    '''
    Function to determine in and out of sample testing of arima model    
    
    arguments
    ---------
    series (seris): time series input
    horizon (int): how far in advance forecast is needed
    start (int): starting location in series
    step_size (int): how often to recalculate forecast
    order (tuple): (p,d,q) order of the model
    seasonal_order (tuple): (P,D,Q,s) seasonal order of model
    
    Returns
    -------
    DataFrame: gives fcst and actuals with date of prediction
    '''
    fcst = []
    actual = []
    date = []
    for i in range(start,len(series)-horizon,step_size):
        model = sm.tsa.statespace.SARIMAX(series[:i+1], #only using data through to and including start 
                                order=order, 
                                seasonal_order=seasonal_order, 
                                trend=trend).fit()
        fcst.append(model.forecast(steps = horizon)[-1]) #forecasting horizon steps into the future
        actual.append(series[i+horizon]) # comparing that to actual value at that point
        date.append(series.index[i+horizon]) # saving date of that value
    return pd.DataFrame({'fcst':fcst,'actual':actual},index=date)
```

```python
#cross validate
warnings.filterwarnings("ignore")
series = data['log_ss']
horizon = 3
start = int(len(data.value)*.75)
step_size = 1
order = (2,0,0)
seasonal_order = (1,1,0,11)

log_cv2 = cross_validate(series,horizon,start,step_size,
                    order = order,
                    seasonal_order = seasonal_order)
```

```python
log_cv2 = np.expm1(log_cv2)
log_cv2.plot()
```
